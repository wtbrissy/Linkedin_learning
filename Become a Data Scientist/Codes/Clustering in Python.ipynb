```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage 
%matplotlib inline
```

## Data 
### Read CSV file

```python
states = pd.read_csv('ClusterData.csv')
```

```python
list(states.columns.values)
```

```python
states
```

### Save numerical data only

```python
st = states[states.columns[2:]]
st.index = states.loc[:,'state_code']
```

## Analysis 
### Creat Linkage matrix for clustring 

```python
z = linkage(st, 'ward')
z
```

### Plot Dendrogram of Clusters

```python
plt.figure(figsize = (25,10))
plt.title('Cluster with All Searches and Personality')
plt.ylabel('Distance')
dendrogram(
    z, 
    labels = st.index,
    leaf_rotation = 0.,
    leaf_font_size = 18.,
)
plt.show()
```
